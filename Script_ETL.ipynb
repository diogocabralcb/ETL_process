```python
import pandas as pd
import csv
```

Problema: O arquivo "T_FATURAS" tinha como separador de números decimais uma vírgula ao invés de um ponto, o que causaria a interpretação de mais colunas que o desejado, já que o separador de colunas é a vírgula.

```python
# Transformando a extensão de um arquivo que estava em .xlsx para .csv

read_file = pd.read_excel ("C:/Users/diogo/Desktop/CASEBI _2/T_CANAIS_DIGITAIS.xlsx") 
read_file.to_csv ("C:/Users/diogo/Desktop/CASEBI _2/T_CANAIS_DIGITAIS.csv",  
                  index = None, 
                  header=True) 
```

```python
# Renomeando o cabeçalho (header) do arquivo

file_name = "C:/Users/diogo/Desktop/CASEBI _2/T_FATURAS.csv"
df = pd.read_csv(file_name, sep=',', names=['DT_VENCIMENTO','DtCorte','MÊS_FATURAMENTO','DS_TIPO_FATURA','ds_rolagem','vl_creditos','VL_FATURA_ANTERIOR','VL_FATURA_ATUAL','NU_FATURA','id_cliente','a','b','c'], header=0)
```

```python
# Concatenação das colunas de valores decimais separados por "," e substituição por "." , através da criação de colunas temporárias

df["vl_creditos_Temp"] = df["vl_creditos"].astype(str) + '.' + df["VL_FATURA_ANTERIOR"].astype(str)
df.head()
df["VL_FATURA_ANTERIOR_Temp"] = df["VL_FATURA_ATUAL"].astype(str) + '.' + df["NU_FATURA"].astype(str)
df["VL_FATURA_ATUAL_Temp"] = df["id_cliente"].astype(str) + '.' + df["a"].astype(str)
df["NU_FATURA_Temp"] = df["b"]
df["id_cliente_Temp"] = df["c"]
```

```python
# Substituição dos valores corrigidos nas colunas originais e drop das colunas temporárias

df["vl_creditos"] = df["vl_creditos_Temp"]
df["VL_FATURA_ANTERIOR"] = df["VL_FATURA_ANTERIOR_Temp"]
df["VL_FATURA_ATUAL"] = df["VL_FATURA_ATUAL_Temp"]
df["NU_FATURA"] = df["NU_FATURA_Temp"]
df["id_cliente"] = df["id_cliente_Temp"]
df.drop(columns=["a", "b","c","vl_creditos_Temp","VL_FATURA_ANTERIOR_Temp","VL_FATURA_ATUAL_Temp","NU_FATURA_Temp","id_cliente_Temp"],inplace=True)
```

```python
# Save e substituição do arquivo

df.to_csv(file_name, encoding='utf-8', header='true', index=False)
```
